```python
import re
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

#读取user_pay_count
user_pay_count_df = pd.read_csv("./data/dataset/dataset/user_pay_count.txt", sep='\t', encoding='UTF-8',parse_dates=True)

import datetime
user_pay_count_df.index = user_pay_count_df['shop_id'];
user_pay_count_df=user_pay_count_df.drop('shop_id',axis=1);
temp = user_pay_count_df.columns[user_pay_count_df.columns != 'shop_id'];
temp = temp.map(lambda str_date:datetime.datetime.strptime(str_date, "%Y-%m-%d %H:%M:%S"))
```

```python
#计算从哪一天开始营业
def findFirstOpenDayIndex(series):
    ##遍历每行，找到第一个不等于0的位置pos 且pos 不大于等于11880 那么 day_index = pos - pos % 24
#     print(series.shape[0])
#     print(user_pay_count_df.shape[1])
    for i in range(series.shape[0]):
        if series[i] != 0:
            if i < user_pay_count_df.shape[1] - 1:
                print('%d:%s'%(series.name, user_pay_count_df.columns[i - i % 24]))
                return (user_pay_count_df.columns[i - i % 24])
            else:
                print("%d no data"%i)
                return user_pay_count_df.columns[i];
```
