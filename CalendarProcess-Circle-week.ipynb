```python
import numpy as np
import pandas as pd
from datetime import datetime
import datetime
```

```python
dates = pd.date_range(start='7/1/2016', end='11/14/2016', freq='D', normalize=True)
s = pd.Series((dates[i].dayofweek for i in range(len(dates))), index=dates)
```

```python
s.name = 'daytype'
```

```python
day_type_df = s.apply(lambda day: day + 1)
```

```python
#1是第一个工作日
#2是第二个工作日
#3是第三个工作日 
#4是第四个工作日 
#5是工作日最后一天 
#6是假期第一天 
#7是假期最后一天
#8是假期中，
```

```python
#假日修正
day_type_df['2016-09-14'] = 5

day_type_df['2016-09-15'] = 6
day_type_df['2016-09-16'] = 8
day_type_df['2016-09-17'] = 7
day_type_df['2016-09-18'] = 1

day_type_df['2016-10-02'] = 8
day_type_df['2016-10-03'] = 8
day_type_df['2016-10-04'] = 8
day_type_df['2016-10-05'] = 8
day_type_df['2016-10-06'] = 8
day_type_df['2016-10-07'] = 7
day_type_df['2016-10-08'] = 1
day_type_df['2016-10-09'] = 1

```

```python
day_type_df = day_type_df.to_frame()
```
