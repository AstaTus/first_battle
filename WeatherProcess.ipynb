```python
def readWeatherData(city):
    weather_info = pd.read_csv("./data/dataset/dataset/city_weather/" + city['pinyin'], 
                               names=['max','min','desc','wind_dir','wind_level'],
                               index_col=0, header=None, encoding='UTF-8',parse_dates=True, dtype={'desc':object, 'wind_level':object})
    return weather_info
```

```python
import string

def changeWeatherDesc(desc):
    if (desc.find('暴雨') != -1 or desc.find('暴雪') != -1 or 
        desc.find('雷阵雨') != -1 or desc.find('雷雨') != -1):
        return 5;
    elif (desc.find('中雨') != -1 or desc.find('大雨') != -1 or 
          desc.find('中雪') != -1 or desc.find('大雪') != -1 or desc.find('雨夹雪') != -1):
        return 4;
    elif(desc.find('阵雪') != -1 or desc.find('阵雨') != -1):
        return 3;
    elif (desc.find('小雨') != -1 or desc.find('小雪') != -1 or desc.find('霾') != -1):
        return 2;
    elif (desc.find('多云') != -1 or desc.find('晴') != -1 or desc.find('阴') != -1 or desc.find('雾') != -1):
        return 1;
    else:
        print("desc" + desc)
        return 0;

def changeWeatherWindLevel(level):
    if (level.find('7') != -1 or level.find('8') != -1 or level.find('8') != -1):
        return 3;
    elif (level.find('4') != -1 or level.find('5') != -1 or level.find('6') != -1):
        return 2;
    elif (level.find('微风') != -1 or level.find('1') != -1 or 
          level.find('2') != -1 or level.find('3') != -1):
        return 1;
    else:
        print("wind" + level)
        return 1;
```

```python
part_city_names = city_names.drop(['思茅'])
```

```python
#city_name = '重庆'

for city_name in part_city_names.index.values:
    print(city_name)
    weather = readWeatherData(city_names.loc[city_name])
    weather['desc'] = weather['desc'].apply(changeWeatherDesc)
    weather['wind_level'] = weather['wind_level'].astype(str)
    weather['wind_level'] = weather['wind_level'].apply(changeWeatherWindLevel)
    time_index=pd.date_range(start='7/1/2015', end='11/30/2016', freq='D')
    weather_df = pd.DataFrame(data=None, index=time_index, columns=['max','desc','wind_level'], dtype=None, copy=False)
    weather_df.index.name = weather.index.name = 'date'
    weather_df['max'] = weather['max']
    weather_df['min'] = weather['min']
    weather_df['desc'] = weather['desc']
    weather_df['wind_level'] = weather['wind_level']
    
    temp_weather = weather_df.reset_index(drop=True)
    #print(weather_df)
    null_indexs = temp_weather[temp_weather['max'].isnull()].index.values;
    print(null_indexs)
    for i in null_indexs:
        if i > 0:
            temp_weather.ix[i] = temp_weather.ix[i - 1];
        else:
            temp_weather.ix[i] = temp_weather.ix[i + 1];
        #print(temp_weather.ix[i])
       
    temp_weather = temp_weather.set_index(weather_df.index);
    temp_weather = temp_weather.astype(np.int32)
    #print(temp_weather)
    #weather = weather.drop('wind_dir', axis = 1)
    temp_weather.to_csv("./data/dataset/dataset/weather/process_" + city_names.loc[city_name]['pinyin'], index=True, encoding='UTF-8')
    
```
