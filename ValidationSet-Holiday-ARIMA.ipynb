```python
#读取预测数据，并且按照时间先后重新排序
pred_datetype_df = holiday_df.reset_index(drop=True).T
arima_df = pd.read_csv("./data/dataset/dataset/holiday_arima/result", encoding='UTF-8', index_col=0)
arima_df.columns = range(0,14)
pred_df = pd.DataFrame(columns=range(0,14))
current_index = 0
for daytype in range(1,6):
    for i in range(0,pred_datetype_df.size):
        if holiday_df[i] == daytype:
            pred_df[i] = arima_df[current_index]
            current_index += 1
```

```python
evaluations = {}
for id in range(1,2001):
    y_arima = pred_df.ix[id].values
    y_Test = base.getTestCount(user_pay_counts, id, pd.to_datetime('2016-10-18'), pd.to_datetime('2016-10-31')).values
    e = base.evaluation(y_Test, y_arima)
    print("id=%d evaluation=%f"%(id, e))
    evaluations[id] = e
```

```python
arima_evaluation_df = pd.DataFrame.from_dict(evaluations, orient='index');
arima_evaluation_df.columns = ['evaluation']
arima_evaluation_df.index.name = 'id'
```

```python
arima_evaluation_df[arima_evaluation_df['evaluation'].isnull()]
arima_evaluation_df = arima_evaluation_df.fillna(1)
```
